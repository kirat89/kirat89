# Cleansing data


```python
#dropping casenum cloumn
claimants.drop(["CASENUM"],axis=1,inplace=True) 
```

```python
claimants.head()
```

```python
claimants.mean()
```

```python
#different values appearing in CLMSEX cloumn & theie count
print(claimants.CLMSEX.value_counts())
#different values appearing in CLMINSUR cloumn & theie count
print(claimants.CLMINSUR.value_counts())
#different values appearing in SEATBELT cloumn & theie count
print(claimants.SEATBELT.value_counts())
#most appearng value in both column
print(claimants.ATTORNEY.value_counts().index[0])
print(claimants.CLMSEX.value_counts().index[0])
print(claimants.SEATBELT.value_counts().index[0])
```

```python
#filng null values
claimants.CLMSEX = claimants.CLMSEX.fillna(claimants.CLMSEX.value_counts().index[0])
claimants.CLMINSUR = claimants.CLMINSUR.fillna(claimants.CLMINSUR.value_counts().index[0])
claimants.SEATBELT = claimants.SEATBELT.fillna(claimants.SEATBELT.value_counts().index[0])
claimants.CLMAGE = claimants.CLMAGE.fillna(claimants.CLMAGE.mean())
```

```python
claimants.isnull().sum()
```

# Data preprocessing


```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(claimants,test_size=0.2)
train
```

```python
x_train=train.iloc[:,1:6 ]
y_train=train.iloc[:,:1]
```

```python
x_test=test.iloc[:,1:6 ]
y_test=test.iloc[:,:1]
```

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(x_train)
logit_model=sm.Logit(y_train,x_train).fit()
logit_model.summary()
```

```python
#predicting for train values
log_train_pred_value=logit_model.predict(x_train)
log_train_pred_value
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})

round(log_train_pred_value)
```

```python
train['Log_pred_value']=round(log_train_pred_value)
train
```

```python
confusion_matrix = pd.crosstab(train['ATTORNEY'],train.Log_pred_value)
confusion_matrix
```

```python
log_train_accuracy=(347+413)/(347+203+109+413)
print(str(log_train_accuracy*100)+' %')
```

```python
#predicting for test values
log_test_pred_value=logit_model.predict(x_test)
log_test_pred_value
```

```python
round(log_test_pred_value)
```

```python
test['Log_pred_value']=round(log_test_pred_value)
test
```

```python
confusion_matrix = pd.crosstab(test['ATTORNEY'],test.Log_pred_value)
confusion_matrix
```

```python
log_test_accuracy=(80+101)/(80+101+55+32)
print(str(log_test_accuracy*100)+' %')
```
