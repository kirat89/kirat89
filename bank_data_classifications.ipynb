# Data preprocessing


```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(bank_data,test_size=0.2)
train
```

```python
x_train=train.iloc[:,:31 ]
y_train=train.iloc[:,31]
```

```python
x_test=test.iloc[:,:31 ]
y_test=test.iloc[:,31]
```

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(x_train)
logit_model=sm.Logit(y_train,x_train).fit()
logit_model.summary()
```

```python
#predicting for train values
log_train_pred_value=logit_model.predict(x_train)
log_train_pred_value
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})

round(log_train_pred_value)
```

```python
train['Log_pred_value']=round(log_train_pred_value)
train
```

```python
confusion_matrix = pd.crosstab(train['y'],train.Log_pred_value)
confusion_matrix
```

```python
log_train_accuracy=(31214+1370)/(31214+729+2855+1370)
print(str(log_train_accuracy*100)+' %')
```

```python
#predicting for test values
log_test_pred_value=logit_model.predict(x_test)
log_test_pred_value
```

```python
round(log_test_pred_value)
```

```python
test['Log_pred_value']=round(log_test_pred_value)
test
```

```python
confusion_matrix = pd.crosstab(test['y'],test.Log_pred_value)
confusion_matrix
```

```python
log_test_accuracy=(7788+346)/(7788+191+718+346)
print(str(log_test_accuracy*100)+' %')
```

```python
Accuracy=[['Logit_model',90.09,89.94]]
```
